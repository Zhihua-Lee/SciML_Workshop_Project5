```python
import numpy as np
from numpy.linalg import svd, solve
import scipy.linalg
from scipy.linalg import sqrtm

def score_enkf_step(
                    x_prior_ensemble, H_func, y, sig_Q, sig_R, is_perturbing_y, iid_perturbating_y=False, 
                    score_enhanced_flag='0', gamma=1, score_function=None, score_prime=None,
                    alternative=False, Cxx_scaling=False
                   ):
    """
    Performs a single iteration step of the Ensemble Kalman Filter, with optional SVD optimization.

    Parameters:
    - x_prior_ensemble: Current ensemble of parameters (N_en x n_x, here n_x=2)
    - H_func: Function handle for the observation operator (maps N_en x n_x -> N_en x n_y, here n_y=1)
    - y: Observation data (scalar or 1-element array)
    - sig_Q: Standard deviation of process noise (scalar, applied element-wise)
    - sig_R: Standard deviation of observation noise (scalar)
    - is_perturbing_y: Boolean flag indicating if observation noise should be added to y
    - score_enhanced_flag: '0' (standard), '1' (explicit score), '2' (implicit score)
    - gamma: Scaling factor for score correction
    - score_function: Function handle for score (gradient). Expected to map
                      (N_en, n_x) ensemble -> (N_en, n_x) score array.
    - score_prime: Function handle for score prime (Hessian). Expected to map
                   (N_en, n_x) ensemble -> list of N_en matrices, each (n_x, n_x).

    Returns:
    - x_prior_ensemble: Updated ensemble of parameters after one iteration
    - x_sample_mean: Mean of the ensemble
    """
    n_en, n_x = x_prior_ensemble.shape

    # # --- Calculate Ensemble Statistics ---
    # x_sample_mean = np.mean(x_prior_ensemble, axis=0) # Shape (n_x,)
    # x_vec = x_prior_ensemble - x_sample_mean          # Shape (N_en, n_x)
    # Cov_xx = (x_vec.T @ x_vec) / (n_en - 1)           # Shape (n_x, n_x)
    
    # Prediction Step: Adding noise to the ensemble
    noise = np.random.normal(0, sig_Q, (n_en, n_x))
    # x_prior_ensemble = x_prior_ensemble + (sqrtm(Cov_xx) @ noise.T).T
    x_prior_ensemble = x_prior_ensemble + ( noise.T).T
    
    # Prediction Step: Adding noise to the ensemble
    # noise = np.random.normal(0, sig_Q, (n_en, n_x))
    # x_prior_ensemble = x_sample_mean + (sqrtm(Cov_xx) @ noise.T).T

    # --- Calculate Ensemble Statistics ---
    x_sample_mean = np.mean(x_prior_ensemble, axis=0) # Shape (n_x,)
    x_vec = x_prior_ensemble - x_sample_mean          # Shape (N_en, n_x)
    Cov_xx = (x_vec.T @ x_vec) / (n_en - 1)           # Shape (n_x, n_x)

    # --- Simulate Observations & Stats ---
    y_sim = H_func(x_prior_ensemble)                  # Shape (N_en, n_y), here n_y=1
    n_y = y_sim.shape[1]
    y_sample_mean = np.mean(y_sim, axis=0)            # Shape (n_y,)
    y_vec = y_sim - y_sample_mean                     # Shape (N_en, n_y)
    

    Cov_yy = (y_vec.T @ y_vec) / (n_en - 1)           # Shape (n_y, n_y)
    obs_err_val = np.linalg.norm(y_sim - y, axis=0) / n_en
    y_sim_max_abs_val = np.max(np.abs(y_sim))
    print(f"observation error: {obs_err_val[0]:.3g}\t y_sim var: {Cov_yy[0][0]:.3g}\t theta var: {Cov_xx[0][0]:.3g}\t y_sim max_abs: {y_sim_max_abs_val:.3g}")
    # --- Cross-Covariance ---
    Cov_xy = (x_vec.T @ y_vec) / (n_en - 1)           # Shape (n_x, n_y)

    # --- Observation Perturbation ---
    # Ensure y is treated as a 1D array for consistency
    y_arr = np.array(y).flatten()
    if is_perturbing_y:
        # Add scalar noise for scalar observation
        y_obs = y_arr + np.random.normal(0, sig_R, y_arr.shape)
    else:
        y_obs = y_arr
    # y_obs shape is now (n_y,)

    # --- Kalman Update (Non-SVD) ---
    S = Cov_yy + (sig_R**2) * np.eye(n_y)             # Shape (n_y, n_y)
    # Solve S K^T = Cov_yx  (where Cov_yx = Cov_xy.T)
    # K^T = S^-1 Cov_yx
    K_transpose = solve(S, Cov_xy.T)                  # Shape (n_y, n_x)
    K = K_transpose.T                                 # Shape (n_x, n_y)
    innovation = y_obs - y_sim                        # Shape (N_en, n_y)
    # (n_x, n_y) @ (n_y, N_en) -> (n_x, N_en) -> Transpose -> (N_en, n_x)
    update = K @ innovation.T # N_x * N_ens
    update = update.T
    
    # Calculate posterior covariance estimate needed for score updates
    C_post = Cov_xx - K @ Cov_xy.T                    # Shape (n_x, n_x)

    match score_enhanced_flag:
        case '0': # Standard EnKF
            pass
        case '1': # Explicit Score-Enhanced EnKF
            # score_update: gamma * (Cov_xx - K @ Cov_yx) @ Cov_xx^-1 @ s_i for each i
            if alternative:
                x_post_ensemble_sim = x_prior_ensemble + update
                score_values = score_function(x_post_ensemble_sim) # Shape (N_en, n_x)
            else:
                score_values = score_function(x_prior_ensemble) # Shape (N_en, n_x)
            scaled_scores = score_values;
            if Cxx_scaling:
                scaled_scores = np.linalg.solve(Cov_xx, score_values.T).T # Shape (N_en, n_x)
            score_update = gamma * (C_post @ scaled_scores.T).T # Shape (N_en, n_x)
            # --- Optional Clipping ---
            # print("max_ensemble_theta:",np.max(np.absolute(x_prior_ensemble)), "\t max_score:", np.max(np.absolute(score_update)))
            # print("score_update \n", score_update[0:3], "\n total update\n",update[0:3])
            clip_threshold = 20.0 # Example threshold
            # score_update = np.clip(score_correction, -clip_threshold, clip_threshold)
            score_update[np.absolute(score_update)>clip_threshold] = 0
            update = update + score_update
        case '2': # Implicit Score-Enhanced EnKF
            # Calculate raw scores and their derivatives (diagonals of Jacobian)
            # ??? jacobian of ensemble or the mean?
            if alternative:
                x_post_ensemble_sim = x_prior_ensemble + update
                score_values = score_function(x_post_ensemble_sim) # Shape (N_en, n_x)
                score_primes = score_prime(x_post_ensemble_sim) # Shape (N_en, n_x , n_x)
            else:
                score_values = score_function(x_prior_ensemble) # Shape (N_en, n_x)
                score_primes = score_prime(x_prior_ensemble) # Shape (N_en, n_x , n_x)
            # Calculate scaled scores z_i = Cov_xx^-1 @ s_i using solve (like case '1')
            # Solves Cov_xx @ Z = score_values.T ; Z has shape (n_x, N_en)
            scaled_scores = score_values;
            if Cxx_scaling:
                scaled_scores = np.linalg.solve(Cov_xx, score_values.T).T # Shape (N_en, n_x)
        
            # Calculate the right-hand side (RHS) vector for each member
            # RHS = update + gamma * C_post @ z_i (vectorized)
            rhs = update + gamma * (C_post @ scaled_scores.T).T # Shape (N_en, n_x)
        
            # --- Loop to solve the linear system for each member ---
            # System: (I - gamma * C_post @ diag(nabla_s_i)) @ delta_x_i = rhs_i
            identity_matrix = np.eye(n_x)                 # Shape (n_x, n_x)
            update = np.zeros_like(x_prior_ensemble)
            for i in range(n_en):
                # Construct the Jacobian matrix for member i
                scaled_score_prime_i = score_primes[i]
                if Cxx_scaling:
                    scaled_score_prime_i = np.linalg.solve(Cov_xx, score_primes[i]) # Shape (n_x, n_x) # Shape (n_x, n_x)
                system_matrix = identity_matrix - gamma * (C_post @ scaled_score_prime_i) # Shape (n_x, n_x)
                update[i] = np.linalg.solve(system_matrix, rhs[i]) # Shape (N_en, n_x)
        case '3':
            # Calculate raw scores and their derivatives (diagonals of Jacobian)
            # ??? jacobian of ensemble or the mean?
            score_values = score_function(x_prior_ensemble)       # Shape (N_en, n_x)
            mean_score_prime = score_prime(x_sample_mean).reshape(1,1) # Shape (n_x , n_x)
            print(mean_score_prime.shape)
            # Calculate scaled scores z_i = Cov_xx^-1 @ s_i using solve (like case '1')
            # Solves Cov_xx @ Z = score_values.T ; Z has shape (n_x, N_en)
            scaled_scores = np.linalg.solve(Cov_xx, score_values.T).T # Shape (N_en, n_x) 
            # Calculate the mean Jacobian across the ensemble
            scaled_mean_score_prime = np.linalg.solve(Cov_xx, mean_score_prime) # Shape (n_x, n_x)
        
            # Calculate the right-hand side (RHS) vector for each member
            # RHS = update + gamma * C_post @ z_i (vectorized)
            rhs = update + gamma * (C_post @ scaled_scores.T).T # Shape (N_en, n_x)
        
            # --- Loop to solve the linear system for each member ---
            # System: (I - gamma * C_post @ diag(nabla_s_i)) @ delta_x_i = rhs_i
            identity_matrix = np.eye(n_x)                 # Shape (n_x, n_x)
            update = np.zeros_like(x_prior_ensemble)
            for i in range(n_en):
                # Construct the Jacobian matrix for member i
                # scaled_score_prime_i = np.linalg.solve(Cov_xx, scaled_mean_score_prime) # Shape (n_x, n_x) # Shape (n_x, n_x)
                system_matrix = identity_matrix - gamma * (C_post @ scaled_mean_score_prime) # Shape (n_x, n_x)
                update[i] = np.linalg.solve(system_matrix, rhs[i]) # Shape (N_en, n_x)
        case '4': # Implicit Score-Enhanced EnKF, statistical linearization for score derivative
            # Calculate raw scores and their derivatives (diagonals of Jacobian)
            # ??? jacobian of ensemble or the mean?
            if alternative:
                x_post_ensemble_sim = x_prior_ensemble + update
                score_values = score_function(x_post_ensemble_sim) # Shape (N_en, n_x)
                # score_primes = score_prime(x_post_ensemble_sim) # Shape (N_en, n_x , n_x)
                x_sample_mean = np.mean(x_post_ensemble_sim, axis=0) # Shape (n_x,)
                x_vec = x_post_ensemble_sim - x_sample_mean          # Shape (N_en, n_x)
                Cov_xx = (x_vec.T @ x_vec) / (n_en - 1)           # Shape (n_x, n_x)
                score_sample_mean = np.mean(score_values, axis=0) # Shape (n_x,)
                score_vec = score_values - score_sample_mean          # Shape (N_en, n_x)
                Cov_xs = (x_vec.T @ score_vec) / (n_en - 1)           # Shape (n_x, n_s)
                score_primes = np.linalg.solve(Cov_xx, Cov_xs).T
            else:
                score_values = score_function(x_prior_ensemble) # Shape (N_en, n_x)
                # score_primes = score_prime(x_prior_ensemble) # Shape (N_en, n_x , n_x)
                # x_sample_mean = np.mean(x_prior_ensemble, axis=0) # Shape (n_x,)
                # x_vec = x_post_ensemble_sim - x_sample_mean          # Shape (N_en, n_x)
                # Cov_xx = (x_vec.T @ x_vec) / (n_en - 1)           # Shape (n_x, n_x)
                score_sample_mean = np.mean(score_values, axis=0) # Shape (n_x,)
                score_vec = score_values - score_sample_mean          # Shape (N_en, n_x)
                Cov_xs = (x_vec.T @ score_vec) / (n_en - 1)           # Shape (n_x, n_s)
                score_primes = np.linalg.solve(Cov_xx, Cov_xs).T
            # Calculate scaled scores z_i = Cov_xx^-1 @ s_i using solve (like case '1')
            # Solves Cov_xx @ Z = score_values.T ; Z has shape (n_x, N_en)
            scaled_scores = score_values;
            if Cxx_scaling:
                scaled_scores = np.linalg.solve(Cov_xx, score_values.T).T # Shape (N_en, n_x)
        
            # Calculate the right-hand side (RHS) vector for each member
            # RHS = update + gamma * C_post @ z_i (vectorized)
            rhs = update + gamma * (C_post @ scaled_scores.T).T # Shape (N_en, n_x)
        
            # --- Loop to solve the linear system for each member ---
            # System: (I - gamma * C_post @ diag(nabla_s_i)) @ delta_x_i = rhs_i
            identity_matrix = np.eye(n_x)                 # Shape (n_x, n_x)
            update = np.zeros_like(x_prior_ensemble)
            for i in range(n_en):
                # Construct the Jacobian matrix for member i
                scaled_score_prime_i = score_primes
                if Cxx_scaling:
                    scaled_score_prime_i = np.linalg.solve(Cov_xx, score_primes) # Shape (n_x, n_x) # Shape (n_x, n_x)
                system_matrix = identity_matrix - gamma * (C_post @ scaled_score_prime_i) # Shape (n_x, n_x)
                update[i] = np.linalg.solve(system_matrix, rhs[i]) # Shape (N_en, n_x)

    
    x_post_ensemble = x_prior_ensemble + update
    x_sample_mean = np.mean(x_post_ensemble, axis=0)

    return x_post_ensemble, x_sample_mean


# def diffusion_eki(y, ensemble, n_iter=100, σ=0.1, γ=0.5, β_max=0.1, τ=20):
#     """扩散增强集合卡尔曼反演"""
#     history = [ensemble.copy()]

#     for k in range(n_iter):
#         # 前向预测
#         G = ensemble**3 - ensemble

#         # 计算协方差
#         mean_θ = np.mean(ensemble)
#         mean_G = np.mean(G)
#         P_θθ = np.mean((ensemble - mean_θ) * (ensemble - mean_θ))
#         P_θG = np.mean((ensemble - mean_θ) * (G - mean_G))
#         P_GG = np.mean((G - mean_G) * (G - mean_G))

#         # 计算分数
#         scores = np.array([score_net(θ, y, σ) for θ in ensemble])
#         mean_s = np.mean(scores)
#         P_θs = np.mean((ensemble - mean_θ) * (scores - mean_s))

#         # 修正的卡尔曼增益
#         K = P_θG / (P_GG + σ**2)

#         # 计算残差
#         residuals = y - G

#         # 分数修正项
#         score_correction = γ * (P_θθ - K * P_θG) * scores

#         # 噪声衰减
#         β_k = β_max * np.exp(-k / τ)
#         noise = np.sqrt(β_k) * np.random.normal(size=ensemble.shape)

        # # 更新集合
        # ensemble += K * residuals + score_correction + noise
        # history.append(ensemble.copy())
```

```python
def score_enkf_process(
    y_obs,
    ensemble,
    H_func,
    method_flag, # Flag: '0' (standard), '1' (explicit score), '2' (implicit score)
    n_iter=50,
    sig_Q=0.1,
    sig_R=0.1,
    is_perturbing_y=True,
    gamma=1.0, # Relevant only for method_flag '1' and '2'
    score_function=None, # Relevant only for method_flag '1' and '2'
    score_prime=None, # Relevant only for method_flag '2'
    alternative=False, Cxx_scaling=False
):
    """
    Unified Ensemble Kalman Filter process supporting standard, explicit score-enhanced,
    and implicit score-enhanced methods via a flag.

    Parameters:
    - y_obs: Observation vector (or scalar) to assimilate.
    - ensemble: Current ensemble (shape: [n_en, n_xi]).
    - H_func: Observation operator function mapping ensemble to simulated observations.
    - method_flag: String flag to select the method:
        '0': Standard EnKF.
        '1': Explicit Score-Enhanced EnKF (requires score_function).
        '2': Implicit Score-Enhanced EnKF (requires score_function and score_prime).
    - n_iter: Number of iterations (default: 50).
    - sig_Q: Standard deviation of process noise (default: 0.1).
    - sig_R: Standard deviation of observation noise (default: 0.1).
    - is_perturbing_y: Boolean flag to decide if y is perturbed (default: True).
    - gamma: Scaling factor for score correction (default: 1.0). Used only if method_flag is '1' or '2'.
    - score_function: Function to compute score. Required if method_flag is '1' or '2'.
    - score_prime: Function to compute score Jacobian/derivative. Required if method_flag is '2'.

    Returns:
    - ensemble: The final updated ensemble.
    - ensemble_history: List recording the ensemble at each iteration.

    Raises:
    - ValueError: If an invalid method_flag is provided.
    - ValueError: If required score functions (score_function for '1',
                   score_function and score_prime for '2') are missing.
    """
    np.random.seed(42) # Consider if seeding here is desired for all use cases

    # --- Input Validation ---
    valid_flags = ['0', '1', '2', '3', '4']
    if method_flag not in valid_flags:
        raise ValueError(f"Invalid method_flag: '{method_flag}'. Must be one of {valid_flags}.")

    if method_flag == '1' and score_function is None:
        raise ValueError("score_function is required when method_flag is '1'.")
    if method_flag == '2':
        if score_function is None:
            raise ValueError("score_function is required when method_flag is '2'.")
        if score_prime is None:
            raise ValueError("score_prime is required when method_flag is '2'.")

    # --- Determine arguments for the step function based on the flag ---
    current_gamma = None
    current_score_function = None
    current_score_prime = None

    if method_flag == '1' or method_flag == '4':
        current_gamma = gamma
        current_score_function = score_function
        # score_prime typically not needed for explicit method in score_enkf_step
    elif method_flag == '2' or method_flag == '3':
        current_gamma = gamma
        current_score_function = score_function
        current_score_prime = score_prime
    # For method_flag '0', all score-related args remain None

    # --- Main Loop ---
    ensemble_history = [ensemble.copy()]
    for k in range(n_iter):
        # Call the underlying step function with determined arguments
        ensemble, ensemble_mean = score_enkf_step(
            ensemble, H_func, y_obs, sig_Q, sig_R, is_perturbing_y, #using_SVD=False,
            score_enhanced_flag=method_flag, # Pass the flag directly
            gamma=current_gamma,
            score_function=current_score_function,
            score_prime=current_score_prime,
            alternative=alternative,
            Cxx_scaling=Cxx_scaling
        )
#         ensemble, ensemble_mean = score_enkf_step(ensemble, H_func, y_obs, sig_Q, sig_R, is_perturbing_y, #using_SVD=False, 
# #                                             score_enhanced_flag='2', gamma=gamma, score_function=score_function, score_prime=score_prime)
#         
        ensemble_history.append(ensemble.copy())

    return ensemble, ensemble_history
```

```python
from scipy.stats import gaussian_kde

def plot_results(standard_history, diffusion_history, theta_true=None, y_obs=None):
    """Visualize the results from Standard EnKF and Diffusion-Enhanced EnKF.
    
    This function creates two figures:
      (1) A 2×2 subplot figure that shows the initial and final distributions for both methods,
      (2) A 1×2 subplot figure that shows the evolution (at selected iterations) of the ensemble density.
    
    All x-axes are forced to have the same range, and all subplots include x-axis labels and tick marks.
    """
    
    # -------- Figure 1: Initial and Final distributions --------
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))  # Removed sharex=True to allow individual control
    x_range = (-2, 2)
    bins = 100
    
    for i, (history, title) in enumerate([
        (standard_history, "Standard EnKF"),
        (diffusion_history, "Diffusion-Enhanced EnKF")
    ]):
        # Initial distribution (top row)
        axes[0, i].hist(history[0], bins=bins, alpha=0.7, density=True)
        axes[0, i].set_title(f"{title} - Initial Distribution")
        axes[0, i].set_xlim(x_range)
        axes[0, i].set_xlabel("θ")
        # Ensure x-ticks are visible
        axes[0, i].tick_params(axis='x', which='both', bottom=True, labelbottom=True)
        
        if theta_true is not None:
            # Accept theta_true as a list of values; if it's not a list, convert it into one.
            if not isinstance(theta_true, (list, tuple, np.ndarray)):
                theta_true = [theta_true]
            for th in theta_true:
                axes[0, i].axvline(th, color='r', linestyle='--', label=f'True θ = {th}')
            axes[0, i].legend()
        
        # Final distribution (bottom row)
        axes[1, i].hist(history[-1], bins=bins, alpha=0.7, density=True)
        axes[1, i].set_title(f"{title} - Final Distribution")
        axes[1, i].set_xlim(x_range)
        axes[1, i].set_xlabel("θ")
        if theta_true is not None:
            # Accept theta_true as a list of values; if it's not a list, convert it into one.
            if not isinstance(theta_true, (list, tuple, np.ndarray)):
                theta_true = [theta_true]
            for th in theta_true:
                axes[1, i].axvline(th, color='r', linestyle='--', label=f'True θ = {th}')
            axes[1, i].legend()
    
    # Make sure x-axis ranges are the same across all subplots
    for ax in axes.flatten():
        ax.set_xlim(x_range)
    
    plt.tight_layout()
    plt.show()
    
    # -------- Figure 2: Evolution of ensemble density over time --------
    fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharex=True)
    x = np.linspace(x_range[0], x_range[1], 1000)
    
    for i, (history, title) in enumerate([
        (standard_history, "Standard EnKF Evolution"),
        (diffusion_history, "Diffusion-Enhanced EnKF Evolution")
    ]):
        # Choose key iterations (evenly spaced: every 1/4 of history length)
        time_points = np.linspace(0, len(history)-1, 4, dtype=int).tolist()

        
        for t in time_points:
            if t < len(history):
                ensemble = history[t]
                try:
                    kde = gaussian_kde(ensemble.flatten())
                    axes[i].plot(x, kde(x), label=f'Iteration {t}')
                except np.linalg.LinAlgError:
                    axes[i].hist(ensemble, bins=bins, alpha=0.3, density=True, label=f'Iteration {t}')
        if theta_true is not None:
            # Accept theta_true as a list of values; if it's not a list, convert it into one.
            if not isinstance(theta_true, (list, tuple, np.ndarray)):
                theta_true = [theta_true]
            for th in theta_true:
                axes[i].axvline(th, color='r', linestyle='--', label=f'True θ = {th}')
        axes[i].set_title(title)
        axes[i].legend()
        axes[i].set_xlabel('θ')
        axes[i].set_ylabel('Density')
        axes[i].set_xlim(x_range)
    
    plt.tight_layout()
    plt.show()

# ------------------ Simple Test ------------------
if __name__ == "__main__":
    # Create simple dummy ensemble histories for testing.
    # For simplicity assume scalar state, using 100 samples per ensemble.
    n_iter = 10       # number of iterations
    n_samples = 100   # number of ensemble members
    standard_history = []
    diffusion_history = []
    
    # Generate a simple evolution: start with a Gaussian centered at 0,
    # and gradually shift the mean.
    for i in range(n_iter):
        standard_history.append(np.random.normal(loc=0.1 * i, scale=0.5, size=n_samples))
        diffusion_history.append(np.random.normal(loc=0.15 * i, scale=0.6, size=n_samples))
    
    theta_true = 1.0  # example true value
    plot_results(standard_history, diffusion_history, theta_true=[0 ,1])
```

```python
def plot_results(histories, titles, theta_true=None, y_obs=None,
                 x_range=(-2, 2), bins=500):
    """
    histories : list  每个元素是一段 EnKF 演化历史 (list of 1-D numpy arrays)
    titles    : list  与 histories 等长，用于子图标题
    theta_true: None、标量或一组标量，用红色虚线标注
    其它参数保持与旧版本一致
    """
    assert len(histories) == len(titles), "histories 与 titles 长度需一致"
    m = len(histories)                     # 方法个数

    # --------- 图 1：初始 & 最终分布（2 × m）---------
    fig1, axes1 = plt.subplots(2, m, figsize=(6*m, 10))

    # 把 theta_true 统一成可迭代
    if theta_true is not None and not isinstance(theta_true, (list, tuple, np.ndarray)):
        theta_true = [theta_true]

    for i, (hist, title) in enumerate(zip(histories, titles)):
        # 初始
        ax_top = axes1[0, i]
        ax_top.hist(hist[0], bins=bins, density=True, alpha=0.7)
        ax_top.set_title(f"{title} - Initial")
        ax_top.set_xlim(x_range)
        ax_top.set_xlabel("θ")

        # 最终
        ax_bot = axes1[1, i]
        ax_bot.hist(hist[-1], bins=bins, density=True, alpha=0.7)
        ax_bot.set_title(f"{title} - Final")
        ax_bot.set_xlim(x_range)
        ax_bot.set_xlabel("θ")

        # 画真值
        if theta_true is not None:
            for th in theta_true:
                ax_top.axvline(th, color='r', linestyle='--')
                ax_bot.axvline(th, color='r', linestyle='--')

    plt.tight_layout()
    plt.show()

    # --------- 图 2：演化过程（1 × m）---------
    fig2, axes2 = plt.subplots(1, m, figsize=(6*m, 6), sharex=True)
    if m == 1:
        axes2 = [axes2]                    # 保持可迭代

    x_plot = np.linspace(*x_range, 1000)

    for i, (hist, title) in enumerate(zip(histories, titles)):
        ax = axes2[i]
        # 选取 4 个等距迭代点（含首尾）
        idxs = np.linspace(0, len(hist)-1, 3, dtype=int)
        for t in idxs:
            ens = hist[t]
            try:
                kde = gaussian_kde(ens.flatten())
                ax.plot(x_plot, kde(x_plot), label=f"Iter {t}")
            except np.linalg.LinAlgError:
                ax.hist(ens, bins=bins, density=True, alpha=0.3, label=f"Iter {t}")

        if theta_true is not None:
            for th in theta_true:
                ax.axvline(th, color='r', linestyle='--')

        ax.set_title(f"{title} Evolution")
        ax.set_xlabel("θ")
        ax.set_ylabel("Density")
        ax.set_xlim(x_range)
        ax.legend()

    plt.tight_layout()
    plt.show()


# ------------------ 简单测试（三条 history） ------------------
if __name__ == "__main__":
    np.random.seed(0)
    n_iter, n_samples = 10, 200

    standard_history  = [np.random.normal(0.1*i, 0.5, n_samples) for i in range(n_iter)]
    diff_exp_history  = [np.random.normal(0.15*i, 0.6, n_samples) for i in range(n_iter)]
    diff_imp_history  = [np.random.normal(0.20*i, 0.7, n_samples) for i in range(n_iter)]

    histories = [standard_history, diff_exp_history, diff_imp_history]
    titles    = ["Standard EnKF", "Diffusion Exp.", "Diffusion Imp."]

    plot_results(histories, titles, theta_true=[-1, 0, 1])
```

```python
from scipy.stats import gaussian_kde

def plot_results(histories, titles, theta_true=None, y_obs=None):
    """Visualize the results from multiple EnKF methods.
    
    This function creates two figures:
      (1) A 2×n subplot figure that shows the initial and final distributions for all methods,
      (2) A 1×n subplot figure that shows the evolution (at selected iterations) of the ensemble density.
    
    Parameters:
    -----------
    histories : list of arrays
        A list where each element is the ensemble history for a method
    titles : list of str
        A list of titles corresponding to each method
    theta_true : float, list, or array, optional
        True parameter value(s) to mark with vertical lines
    y_obs : array, optional
        Observation data (not used in current implementation)
    
    All x-axes are forced to have the same range, and all subplots include x-axis labels and tick marks.
    """
    
    n_methods = len(histories)
    
    # -------- Figure 1: Initial and Final distributions --------
    fig, axes = plt.subplots(2, n_methods, figsize=(7*n_methods, 10))  
    x_range = (-2, 2)
    bins = 100
    
    # Handle case where there's only one method (axes won't be 2D)
    if n_methods == 1:
        axes = np.array([[axes[0]], [axes[1]]])
    
    for i, (history, title) in enumerate(zip(histories, titles)):
        # Initial distribution (top row)
        axes[0, i].hist(history[0], bins=bins, alpha=0.7, density=True)
        axes[0, i].set_title(f"{title} - Initial Distribution")
        axes[0, i].set_xlim(x_range)
        axes[0, i].set_xlabel("θ")
        # Ensure x-ticks are visible
        axes[0, i].tick_params(axis='x', which='both', bottom=True, labelbottom=True)
        
        if theta_true is not None:
            # Accept theta_true as a list of values; if it's not a list, convert it into one.
            if not isinstance(theta_true, (list, tuple, np.ndarray)):
                theta_true = [theta_true]
            for th in theta_true:
                axes[0, i].axvline(th, color='r', linestyle='--', label=f'True θ = {th}')
            axes[0, i].legend()
        
        # Final distribution (bottom row)
        axes[1, i].hist(history[-1], bins=bins, alpha=0.7, density=True)
        axes[1, i].set_title(f"{title} - Final Distribution")
        axes[1, i].set_xlim(x_range)
        axes[1, i].set_xlabel("θ")
        if theta_true is not None:
            # Accept theta_true as a list of values; if it's not a list, convert it into one.
            if not isinstance(theta_true, (list, tuple, np.ndarray)):
                theta_true = [theta_true]
            for th in theta_true:
                axes[1, i].axvline(th, color='r', linestyle='--', label=f'True θ = {th}')
            axes[1, i].legend()
    
    # Make sure x-axis ranges are the same across all subplots
    for ax in axes.flatten():
        ax.set_xlim(x_range)
    
    plt.tight_layout()
    plt.show()
    
    # -------- Figure 2: Evolution of ensemble density over time --------
    fig, axes = plt.subplots(1, n_methods, figsize=(8*n_methods, 6), sharex=True)
    
    # Handle case where there's only one method (axes won't be array)
    if n_methods == 1:
        axes = np.array([axes])
        
    x = np.linspace(x_range[0], x_range[1], 1000)
    
    for i, (history, title) in enumerate(zip(histories, titles)):
        # Choose key iterations (evenly spaced: every 1/4 of history length)
        time_points = np.linspace(0, len(history)-1, 4, dtype=int).tolist()
        
        for t in time_points:
            if t < len(history):
                ensemble = history[t]
                try:
                    kde = gaussian_kde(ensemble.flatten())
                    axes[i].plot(x, kde(x), label=f'Iteration {t}')
                except np.linalg.LinAlgError:
                    axes[i].hist(ensemble, bins=bins, alpha=0.3, density=True, label=f'Iteration {t}')
        if theta_true is not None:
            # Accept theta_true as a list of values; if it's not a list, convert it into one.
            if not isinstance(theta_true, (list, tuple, np.ndarray)):
                theta_true = [theta_true]
            for th in theta_true:
                axes[i].axvline(th, color='r', linestyle='--', label=f'True θ = {th}')
        axes[i].set_title(title)
        axes[i].legend()
        axes[i].set_xlabel('θ')
        axes[i].set_ylabel('Density')
        axes[i].set_xlim(x_range)
    
    plt.tight_layout()
    plt.show()

# Example usage:
if __name__ == "__main__":
    # Create simple dummy ensemble histories for testing.
    n_iter = 10       # number of iterations
    n_samples = 100   # number of ensemble members
    
    # For two methods (original case)
    standard_history = []
    diffusion_history = []
    
    # Generate a simple evolution: start with a Gaussian centered at 0,
    # and gradually shift the mean.
    for i in range(n_iter):
        standard_history.append(np.random.normal(loc=0.1 * i, scale=0.5, size=n_samples))
        diffusion_history.append(np.random.normal(loc=0.15 * i, scale=0.6, size=n_samples))
    
    theta_true = 1.0  # example true value
    
    # Use the new interface
    plot_results(
        histories=[standard_history, diffusion_history],
        titles=["Standard EnKF", "Diffusion-Enhanced EnKF"],
        theta_true=[0, 1]
    )
```

## define forward model and score function

```python
import sympy as sp
import numpy as np
from IPython.display import display, Math

# Initialize SymPy for LaTeX printing in environments like Jupyter or Colab
sp.init_printing(use_latex="mathjax")

# === 1. Define Symbolic Expressions ===

# Define symbolic variables
theta = sp.Symbol("theta", real=True)          # Parameter of interest
y = sp.Symbol("y", real=True)              # Observed data
sigma = sp.Symbol("sigma", positive=True)  # Standard deviation (must be > 0)

# Forward model: f(theta) = theta^3 - theta
f = theta**3 - theta

# Define components of the Bayesian model (log-probabilities, unnormalized)
# Log Unnormalized Prior: log p(theta)
log_unnorm_prior = -0.5 * theta**2
# Log Unnormalized Likelihood: log p(y | theta)
log_unnorm_likelihood = -0.5 * ((y - f) / sigma)**2

# Log Unnormalized Posterior: log p(theta | y) = log p(theta) + log p(y | theta) (+ const)
log_unnorm_posterior = log_unnorm_prior + log_unnorm_likelihood

# Derivatives of the log-posterior w.r.t. theta
# Score function (1st derivative)
score_function = sp.diff(log_unnorm_posterior, theta)
# Score prime (2nd derivative)
score_prime = sp.diff(score_function, theta)


# === 2. Export Symbolic Expressions to NumPy Functions ===

# Use sympy.lambdify to create callable functions for numerical computation
f_np = sp.lambdify((theta,), f, "numpy")
log_unnorm_prior_np = sp.lambdify((theta,), log_unnorm_prior, "numpy")
log_unnorm_likelihood_np = sp.lambdify((theta, y, sigma), log_unnorm_likelihood, "numpy")
log_unnorm_posterior_np = sp.lambdify((theta, y, sigma), log_unnorm_posterior, "numpy")
score_np = sp.lambdify((theta, y, sigma), score_function, "numpy")
score_prime_np = sp.lambdify((theta, y, sigma), score_prime, "numpy")


# === 3. Display Simplified Symbolic Formulas ===

print("=== Simplified Symbolic Definitions (LaTeX Formulas) ===")

print("\nForward model f:")
# Simplify expressions before displaying using LaTeX
display(Math(f"f({sp.latex(theta)}) = {sp.latex(sp.simplify(f))}"))
print("\nLog Unnormalized Prior:")
display(Math(f"\\log p({sp.latex(theta)}) \\propto {sp.latex(sp.simplify(log_unnorm_prior))}"))
print("\nLog Unnormalized Likelihood:")
display(Math(f"\\log p(y | {sp.latex(theta)}) \\propto {sp.latex(sp.simplify(log_unnorm_likelihood))}"))
print("\nLog Unnormalized Posterior:")
display(Math(f"\\log p({sp.latex(theta)} | y) \\propto {sp.latex(sp.simplify(log_unnorm_posterior))}"))
print("\nScore Function (Derivative):")
# Display the simplified first derivative
display(Math(f"\\frac{{\\partial}}{{\\partial {sp.latex(theta)}}} \\log p({sp.latex(theta)} | y) = {sp.latex(sp.simplify(score_function))}"))
print("\nScore Prime (Second Derivative):")
# Display the simplified second derivative
display(Math(f"\\frac{{\\partial^2}}{{\\partial {sp.latex(theta)}^2}} \\log p({sp.latex(theta)} | y) = {sp.latex(sp.simplify(score_prime))}"))
```

```python
np.random.seed(42)

def run_unimodal_test(gamma = 0.01):
    """Run a special multimodal test case (for y=0, posterior ~ double-peaked at θ≈-1,1)."""

    # Choose y=0 so that p(θ|y=0) is clearly multimodal at about 0, -1 and +1
    y_val = 1
    sigma_val = 0.1
    y_obs = y_val

    # Simple helper that provides a score(θ) for a fixed y,σ
    def score_given_y_sigma(th):
        return score_np(th, y_val, sigma_val)
    def score_prime_given_y_sigma(th):
        return score_prime_np(th, y_val, sigma_val)

    # Parameters for EnKF
    n_xi = 1
    n_en = 10000         # number of ensemble members
    n_iter = 50        # number of iterations
    sig_Q = 0.01        # process noise std
    sig_R = 0.01        # observation noise std
    is_perturbing_y = True
    P = 1**2 * np.eye(n_xi)  # initial covariance

    gamma = gamma

    alternative=True
    Cxx_scaling=True

    # Initial ensemble: draw from N(0,P)
    m_0 = np.zeros(n_xi)  
    x_prior_ensemble = m_0 + np.random.multivariate_normal(np.zeros(n_xi), P, size=n_en) 
    print("x_prior_ensemble shape:", x_prior_ensemble.shape)


    # Run standard EnKF
    print("\nRunning standard EnKF...")
    standard_ensemble, standard_history = score_enkf_process(
        y_obs=y_obs,
        ensemble=x_prior_ensemble.copy(),
        H_func=f_np,
        n_iter=n_iter,
        sig_Q=sig_Q,
        sig_R=sig_R,
        is_perturbing_y=is_perturbing_y,
        method_flag='0',
    )

    # Run diffusion-enhanced(explicit) EnKF (increase gamma for stronger score guidance)
    print("\nRunning diffusion-enhanced(explicit) EnKF...")
    diffusion_ensemble_explicit, diffusion_history_explicit = score_enkf_process(
        y_obs=y_obs,
        ensemble=x_prior_ensemble.copy(),
        H_func=f_np,
        n_iter=n_iter,
        sig_Q=sig_Q,
        sig_R=sig_R,
        is_perturbing_y=is_perturbing_y,
        method_flag='1',
        gamma=gamma,
        score_function=score_given_y_sigma,
        alternative=alternative,
        Cxx_scaling=Cxx_scaling
    )

    # Run diffusion-enhanced(explicit) EnKF (increase gamma for stronger score guidance)
    print("\nRunning diffusion-enhanced(implicit) EnKF...")
    diffusion_ensemble_implicit, diffusion_history_implicit = score_enkf_process(
        y_obs=y_obs,
        ensemble=x_prior_ensemble.copy(),
        H_func=f_np,
        n_iter=n_iter,
        sig_Q=sig_Q,
        sig_R=sig_R,
        is_perturbing_y=is_perturbing_y,
        method_flag='2',
        gamma=gamma,
        score_function=score_given_y_sigma,
        score_prime=score_prime_given_y_sigma,
        alternative=alternative,
        Cxx_scaling=Cxx_scaling
    )

    # Run diffusion-enhanced(implicit) EnKF with statistical linearization (CASE 4)
    print(f"\nRunning diffusion-enhanced(implicit, stat-lin) EnKF...")
    diffusion_ensemble_statlin, diffusion_history_statlin = score_enkf_process(
        y_obs=y_obs,
        ensemble=x_prior_ensemble.copy(),
        H_func=f_np,
        n_iter=n_iter,
        sig_Q=sig_Q,
        sig_R=sig_R,
        is_perturbing_y=is_perturbing_y,
        method_flag='4', # Use new flag for statistical linearization
        gamma=gamma,
        score_function=score_given_y_sigma, # Needed to get score_values
        score_prime=None, # Not used by method '4' as it calculates Jacobian internally
        alternative=alternative,
        Cxx_scaling=Cxx_scaling
    )

    # --- Visualize results with your plot_results function ---
    histories = [standard_history, diffusion_history_explicit, diffusion_history_implicit, diffusion_history_statlin]
    titles    = ["Standard EnKF", "Explicit Score EnKF", "Implicit Score EnKF (Analytic Prime)", "Implicit Score EnKF (Stat-Lin Prime)"]
    
    theta_true = [1.32]
    plot_results(histories, titles, theta_true=theta_true)

run_unimodal_test(gamma = 0.01)
```

# 2D Two-modal Posterior Problem

This problem is to reproduce the example from https://github.com/PKU-CMEGroup/InverseProblems.jl/blob/master/Multimodal/Twomodal-Density-2D.ipynb.

## Problem Definition

We consider the following inverse problem:

$$y = G(\theta)+ η=(θ_1 - θ_2)^2 + η$$

where:
- $θ = [θ_1, θ_2]$ is the two-dimensional parameter we want to estimate
- $y$ is the observation, and we set $y = 4.2297$
- $η$ is the observation noise, following $\mathcal{N}(0, 1)$
- The prior distribution is $θ \sim \mathcal{N}(\mu_0, I)$, and we will consider two cases: $\mu_0 = [0, 0]^T$ and $\mu_0 = [0.5, 0]^T$

The interesting aspect of this problem is that for a fixed observation $y$, the parameters $θ$ that satisfy the condition form two strip-shaped regions in the parameter space. This is because the solution to the equation $(θ_1 - θ_2)^2 = y$ in two-dimensional space is $ |θ_1 - θ_2| = \sqrt{y} $, which can be further divided into two lines: $\{(\theta_1,\theta_2) \in \mathbb{R}^2| θ_2 - θ_1 = \sqrt{y}\}\cup\{(\theta_1,\theta_2) \in \mathbb{R}^2| θ_2 - θ_1 = -\sqrt{y}\}$.

After considering the noise $\eta$, the posterior distribution spreads and blurs from the above two lines into a bimodal, parallel strip distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import sympy as sp
from IPython.display import display
from sympy import Eq
```

## Define the forward model and scoring function.

```python
import sympy as sp
import numpy as np
from IPython.display import display, Math

# Initialize SymPy for LaTeX printing in environments like Jupyter or Colab
sp.init_printing(use_latex="mathjax")

# === 1. Define Symbolic Expressions (Vector Parameter) ===

# Define symbolic variables
theta = sp.Matrix(sp.symbols('theta1:3'))  # Parameter vector [theta1, theta2]^T
y = sp.Symbol('y', real=True)              # Scalar observation
sigma = sp.Symbol('sigma', positive=True)  # Standard deviation (> 0)
mu0 = sp.Matrix(sp.symbols('mu01:3'))    # Prior mean vector [mu01, mu02]^T

# Forward model: f(theta) (scalar output in this example)
f = (theta[0] - theta[1])**2

# Define components of the Bayesian model (log-probabilities, unnormalized)

# Log Unnormalized Prior: log p(theta)
# Calculated as -1/2 * (theta - mu0)^T * (theta - mu0)
log_unnorm_prior_matrix = -sp.Rational(1, 2) * (theta - mu0).T * (theta - mu0)
# The result is a 1x1 matrix; extract the scalar expression for consistency
log_unnorm_prior = log_unnorm_prior_matrix[0]

# Log Unnormalized Likelihood: log p(y | theta)
log_unnorm_likelihood = -sp.Rational(1, 2) * ((y - f) / sigma)**2

# Log Unnormalized Posterior: log p(theta | y)
log_unnorm_posterior = log_unnorm_prior + log_unnorm_likelihood

# Gradient of log-posterior w.r.t. theta (Score vector)
# Result is a column vector: [d/dtheta1, d/dtheta2]^T
score_function = sp.Matrix([sp.diff(log_unnorm_posterior, t) for t in theta])

# Hessian matrix of log-posterior w.r.t. theta
# Result is a 2x2 matrix of second partial derivatives
hessian_matrix = sp.hessian(log_unnorm_posterior, theta)


# === 2. Export Symbolic Expressions to NumPy Functions ===

# Use sympy.lambdify for numerical evaluation.
# Matrix/vector inputs (like theta, mu0) expect corresponding NumPy arrays.
f_np = sp.lambdify((theta,), f, "numpy")
log_unnorm_prior_np = sp.lambdify((theta, mu0), log_unnorm_prior, "numpy")
log_unnorm_likelihood_np = sp.lambdify((theta, y, sigma), log_unnorm_likelihood, "numpy")
log_unnorm_posterior_np = sp.lambdify((theta, mu0, y, sigma), log_unnorm_posterior, "numpy")
# score_function_np will return a NumPy column vector
score_function_np = sp.lambdify((theta, mu0, y, sigma), score_function, "numpy")
# hessian_matrix_np will return a 2x2 NumPy array
hessian_matrix_np = sp.lambdify((theta, mu0, y, sigma), hessian_matrix, "numpy")


# === 3. Display Simplified Symbolic Formulas ===

print("=== Simplified Symbolic Definitions (LaTeX Formulas) ===")

print("\nParameter vector theta:")
display(theta) # Direct display renders matrices well
print("\nPrior mean vector mu0:")
display(mu0) # Direct display renders matrices well
print("\nForward model f:")
# Simplify expressions before displaying using LaTeX
display(Math(f"f({sp.latex(theta)}) = {sp.latex(sp.simplify(f))}"))
print("\nLog Unnormalized Prior:")
display(Math(f"\\log p({sp.latex(theta)}) \\propto {sp.latex(sp.simplify(log_unnorm_prior))}"))
print("\nLog Unnormalized Likelihood:")
display(Math(f"\\log p(y | {sp.latex(theta)}) \\propto {sp.latex(sp.simplify(log_unnorm_likelihood))}"))
print("\nLog Unnormalized Posterior:")
display(Math(f"\\log p({sp.latex(theta)} | y) \\propto {sp.latex(sp.simplify(log_unnorm_posterior))}"))
print("\nScore Function (Gradient):")
# Simplify applies element-wise to matrices/vectors
display(Math(f"\\nabla_{{{sp.latex(theta)}}} \\log p({sp.latex(theta)} | y) = {sp.latex(sp.simplify(score_function))}"))
print("\nHessian Matrix:")
display(Math(f"\\nabla^2_{{{sp.latex(theta)}}} \\log p({sp.latex(theta)} | y) = {sp.latex(sp.simplify(hessian_matrix))}"))
```

## Plotting Function for 2D Ensembles

Let:
* $p_{ref}(\theta)$ be the reference posterior probability density function.
* $p_{GMKI}(\theta)$ be the approximate posterior density function obtained from GMKI at a specific iteration.
* $\theta_{i,j} = (\theta_{1,i}, \theta_{2,j})$ represent the discrete grid points.
* $\Delta A = \Delta \theta_1 \Delta \theta_2$ be the area of a grid cell (`(xx[2]-xx[1])*(yy[2]-yy[1])`).
* $Z_{ref}[i,j] \approx p_{ref}(\theta_{i,j})$ and $Z_{GMKI}[i,j] \approx p_{GMKI}(\theta_{i,j})$ be the density values on the grid.

The code from https://github.com/PKU-CMEGroup/InverseProblems.jl/blob/master/Multimodal/Twomodal-Density-2D.ipynb calculates the error $E$ using a discrete sum:

$$ E = \sum_{i,j} | Z_{GMKI}[i,j] - Z_{ref}[i,j] | \Delta A $$

This sum approximates the $L_1$ distance between the two density functions:

$$ E \approx \int | p_{GMKI}(\theta) - p_{ref}(\theta) | d\theta $$

The standard **Total Variation (TV) distance** is defined as:

$$ TV(p_{GMKI}, p_{ref}) = \frac{1}{2} \int | p_{GMKI}(\theta) - p_{ref}(\theta) | d\theta $$

Therefore, the value calculated in the code and plotted as "TV distance" is an approximation of **twice the standard Total Variation distance**:

$$ E \approx 2 \times TV(p_{GMKI}, p_{ref}) $$

```python
import matplotlib.pyplot as plt
import numpy as np
# from scipy.stats import gaussian_kde # Optional: for density contours

def plot_results_2d(histories, titles, theta_true=None, mu0=None, param_range=None):
    """
    Visualize the results from multiple EnKF methods for a 2D parameter space.

    Parameters:
    -----------
    histories : list of lists of arrays
        List where each element is the ensemble history for a method.
        Each history is a list of numpy arrays, where each array has shape (n_ensemble, 2).
    titles : list of str
        Titles corresponding to each method.
    theta_true : array-like (len 2), optional
        True parameter vector [theta1_true, theta2_true] to mark.
    mu0 : array-like (len 2), optional
        Prior mean vector [mu01, mu02] to mark.
    param_range : list or tuple (xmin, xmax, ymin, ymax), optional
        Axis limits for the plots. If None, determined automatically.
    """
    n_methods = len(histories)
    assert n_methods == len(titles), "Number of histories and titles must match."

    # Determine plot limits
    if param_range:
        xmin, xmax, ymin, ymax = param_range
    else:
        # Auto-determine range from all ensembles
        all_thetas = np.vstack([item for history in histories for item in history])
        xmin, ymin = np.percentile(all_thetas, 1, axis=0)
        xmax, ymax = np.percentile(all_thetas, 99, axis=0)
        x_margin = (xmax - xmin) * 0.1
        y_margin = (ymax - ymin) * 0.1
        xmin -= x_margin
        xmax += x_margin
        ymin -= y_margin
        ymax += y_margin
        param_range = (xmin, xmax, ymin, ymax)


    # -------- Figure 1: Initial and Final Distributions (Scatter Plots) --------
    fig1, axes1 = plt.subplots(2, n_methods, figsize=(6 * n_methods, 10), sharex=True, sharey=True)
    if n_methods == 1: axes1 = np.array([[axes1[0]], [axes1[1]]]) # Ensure 2D indexing

    for i, (history, title) in enumerate(zip(histories, titles)):
        # Initial ensemble (top row)
        ax_top = axes1[0, i]
        ens_init = history[0]
        ax_top.scatter(ens_init[:, 0], ens_init[:, 1], alpha=0.3, s=5)
        ax_top.set_title(f"{title} - Initial Ensemble")
        ax_top.set_ylabel("θ₂")
        if mu0 is not None:
             ax_top.scatter(mu0[0], mu0[1], color='orange', marker='x', s=100, label='Prior Mean (μ₀)')
        if theta_true is not None:
            ax_top.scatter(theta_true[0], theta_true[1], color='red', marker='*', s=150, label='True θ')
        if mu0 is not None or theta_true is not None:
            ax_top.legend(loc='upper right')
        ax_top.grid(True, linestyle=':')


        # Final ensemble (bottom row)
        ax_bot = axes1[1, i]
        ens_final = history[-1]
        ax_bot.scatter(ens_final[:, 0], ens_final[:, 1], alpha=0.3, s=5)
        ax_bot.set_title(f"{title} - Final Ensemble")
        ax_bot.set_xlabel("θ₁")
        ax_bot.set_ylabel("θ₂")
        if mu0 is not None:
             ax_bot.scatter(mu0[0], mu0[1], color='orange', marker='x', s=100, label='Prior Mean (μ₀)')
        if theta_true is not None:
            ax_bot.scatter(theta_true[0], theta_true[1], color='red', marker='*', s=150, label='True θ')
        if mu0 is not None or theta_true is not None:
            ax_bot.legend(loc='upper right')
        ax_bot.grid(True, linestyle=':')

    # Apply shared limits
    for ax in axes1.flat:
        ax.set_xlim(param_range[0], param_range[1])
        ax.set_ylim(param_range[2], param_range[3])

    plt.tight_layout()
    plt.show()


    # -------- Figure 2: Evolution of Ensemble (Selected Iterations) --------
    fig2, axes2 = plt.subplots(1, n_methods, figsize=(6 * n_methods, 6), sharex=True, sharey=True)
    if n_methods == 1: axes2 = [axes2] # Ensure iterable

    colors = plt.cm.viridis(np.linspace(0, 1, 4)) # Colors for different iterations

    for i, (history, title) in enumerate(zip(histories, titles)):
        ax = axes2[i]
        # Select 4 evenly spaced iterations (including start and end)
        idxs = np.linspace(0, len(history) - 1, 4, dtype=int)

        for k, t in enumerate(idxs):
            ens = history[t]
            ax.scatter(ens[:, 0], ens[:, 1], alpha=0.2 + 0.15*k, s=5, color=colors[k], label=f"Iter {t}")

        if mu0 is not None:
             ax.scatter(mu0[0], mu0[1], color='orange', marker='x', s=100, label='Prior Mean (μ₀)')
        if theta_true is not None:
            ax.scatter(theta_true[0], theta_true[1], color='red', marker='*', s=150, label='True θ')

        ax.set_title(f"{title} Evolution")
        ax.set_xlabel("θ₁")
        ax.set_ylabel("θ₂")
        ax.legend(loc='upper right')
        ax.grid(True, linestyle=':')

    # Apply shared limits
    for ax in axes2:
        ax.set_xlim(param_range[0], param_range[1])
        ax.set_ylim(param_range[2], param_range[3])

    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde # Import KDE
# from functools import partial # May be useful for creating log_posterior_func

def plot_results_2d(
    histories,
    titles,
    theta_true=None,
    mu0=None,
    param_range=None,        # Optional: Limits for scatter plots (xmin, xmax, ymin, ymax)
    # New parameters for TV distance plot using analytical function
    log_posterior_func=None, # Analytical function log p(theta|y) + C (unnormalized is OK), takes theta=[t1,t2] 
    grid_range=None,         # Optional: (xmin, xmax, ymin, ymax) for density grid calculation
    grid_resolution=(100, 100) # Optional: (Nx, Ny) points for density grid
):
    """
    Visualize results from multiple EnKF methods for a 2D parameter space,
    optionally adding a plot for TV distance evolution calculated against
    an analytical reference posterior defined on an internal grid.

    Parameters:
    -----------
    histories : list of lists of arrays
        List of ensemble histories, each history is a list of arrays (n_ensemble, 2).
    titles : list of str
        Titles corresponding to each method.
    theta_true : array-like (len 2), optional
        True parameter vector to mark.
    mu0 : array-like (len 2), optional
        Prior mean vector to mark.
    param_range : list or tuple (xmin, xmax, ymin, ymax), optional
        Axis limits for the scatter/evolution plots. If None, determined automatically
        from ensemble data.
    log_posterior_func : callable, optional
        A function that takes theta (array-like, len 2) and returns the
        log posterior density log p(theta|y). Required for TV plot.
        Example: log_posterior_func(np.array([t1, t2])) -> float
    grid_range : tuple (xmin, xmax, ymin, ymax), optional
        The range over which to define the internal grid for evaluating the
        reference PDF and KDEs for TV distance calculation.
        If None, defaults to `param_range` (either user-provided or auto-calculated).
    grid_resolution : tuple (Nx, Ny), optional
        Number of points along each axis for the density grid (default: (100, 100)).
    """
    n_methods = len(histories)
    assert n_methods == len(titles), "Number of histories and titles must match."

    # --- Determine plot limits for scatter/evolution plots (param_range) ---
    auto_calculated_param_range = False
    if param_range is None:
        auto_calculated_param_range = True
        # Auto-determine range from all ensembles
        all_valid_thetas = [item for history in histories for item in history if item.size > 0]
        if not all_valid_thetas:
             xmin, xmax, ymin, ymax = -1, 1, -1, 1 # Default if no data
             print("Warning: No valid ensemble data. Using default scatter plot range [-1, 1].")
        else:
            all_thetas = np.vstack(all_valid_thetas)
            xmin, ymin = np.percentile(all_thetas, 1, axis=0)
            xmax, ymax = np.percentile(all_thetas, 99, axis=0)
            x_margin = (xmax - xmin) * 0.1 if xmax > xmin else 0.1
            y_margin = (ymax - ymin) * 0.1 if ymax > ymin else 0.1
            xmin -= x_margin
            xmax += x_margin
            ymin -= y_margin
            ymax += y_margin
        param_range = (xmin, xmax, ymin, ymax)
        print(f"Auto-determined param_range for scatter plots: {param_range}")
    else:
         xmin, xmax, ymin, ymax = param_range


    # -------- Figure 1: Initial and Final Distributions (Scatter Plots) --------
    # (Plotting code for Figure 1 remains unchanged, uses param_range)
    fig1, axes1 = plt.subplots(2, n_methods, figsize=(6 * n_methods, 10), sharex=True, sharey=True)
    if n_methods == 1: axes1 = np.array([[axes1[0]], [axes1[1]]]) # Ensure 2D indexing
    for i, (history, title) in enumerate(zip(histories, titles)):
        if not history: continue
        # Initial
        ax_top = axes1[0, i]
        if history[0].size > 0: ax_top.scatter(history[0][:, 0], history[0][:, 1], alpha=0.3, s=5)
        ax_top.set_title(f"{title} - Initial")
        ax_top.set_ylabel("θ₂")
        if mu0 is not None: ax_top.scatter(mu0[0], mu0[1], c='orange', marker='x', s=100, label='μ₀')
        if theta_true is not None: ax_top.scatter(theta_true[0], theta_true[1], c='red', marker='*', s=150, label='True θ')
        if mu0 is not None or theta_true is not None: ax_top.legend(loc='upper right')
        ax_top.grid(True, linestyle=':')
        # Final
        ax_bot = axes1[1, i]
        if history[-1].size > 0: ax_bot.scatter(history[-1][:, 0], history[-1][:, 1], alpha=0.3, s=5)
        ax_bot.set_title(f"{title} - Final")
        ax_bot.set_xlabel("θ₁")
        ax_bot.set_ylabel("θ₂")
        if mu0 is not None: ax_bot.scatter(mu0[0], mu0[1], c='orange', marker='x', s=100, label='μ₀')
        if theta_true is not None: ax_bot.scatter(theta_true[0], theta_true[1], c='red', marker='*', s=150, label='True θ')
        if mu0 is not None or theta_true is not None: ax_bot.legend(loc='upper right')
        ax_bot.grid(True, linestyle=':')
    for ax in axes1.flat:
        ax.set_xlim(param_range[0], param_range[1])
        ax.set_ylim(param_range[2], param_range[3])
        ax.set_aspect('equal', adjustable='box')
    fig1.tight_layout()
    plt.show()


    # -------- Figure 2: Evolution of Ensemble (Selected Iterations) --------
    # (Plotting code for Figure 2 remains unchanged, uses param_range)
    fig2, axes2 = plt.subplots(1, n_methods, figsize=(6 * n_methods, 6), sharex=True, sharey=True)
    if n_methods == 1: axes2 = [axes2] # Ensure iterable
    colors = plt.cm.viridis(np.linspace(0, 1, 4))
    for i, (history, title) in enumerate(zip(histories, titles)):
        if not history: continue
        ax = axes2[i]
        idxs = np.linspace(0, len(history) - 1, 4, dtype=int)
        for k, t in enumerate(idxs):
             if history[t].size > 0:
                 ax.scatter(history[t][:, 0], history[t][:, 1], alpha=0.2 + 0.15*k, s=5, color=colors[k], label=f"Iter {t}")
        if mu0 is not None: ax.scatter(mu0[0], mu0[1], c='orange', marker='x', s=100, label='μ₀')
        if theta_true is not None: ax.scatter(theta_true[0], theta_true[1], c='red', marker='*', s=150, label='True θ')
        ax.set_title(f"{title} Evolution")
        ax.set_xlabel("θ₁")
        ax.set_ylabel("θ₂")
        ax.legend(loc='upper right')
        ax.grid(True, linestyle=':')
    for ax in axes2:
        ax.set_xlim(param_range[0], param_range[1])
        ax.set_ylim(param_range[2], param_range[3])
        ax.set_aspect('equal', adjustable='box')
    fig2.tight_layout()
    plt.show()


    # -------- Figure 3: TV Distance Evolution (Optional) --------
    if log_posterior_func is None:
        print("Analytical log posterior function (log_posterior_func) not provided, skipping TV distance plot.")
        return # Exit function

    # --- Determine grid range and resolution ---
    if grid_range is None:
        grid_range = param_range # Default to scatter plot range if not specified
        print(f"Using param_range {grid_range} for density grid.")
    grid_xmin, grid_xmax, grid_ymin, grid_ymax = grid_range
    Nx, Ny = grid_resolution

    if not all(isinstance(n, int) and n > 1 for n in [Nx, Ny]):
         print(f"Warning: Invalid grid_resolution {grid_resolution}. Nx and Ny must be integers > 1. Skipping TV plot.")
         return

    # --- Generate Grid ---
    xx = np.linspace(grid_xmin, grid_xmax, Nx)
    yy = np.linspace(grid_ymin, grid_ymax, Ny)
    dx = xx[1] - xx[0]
    dy = yy[1] - yy[0]
    T1, T2 = np.meshgrid(xx, yy) # T1 shape (Ny, Nx), T2 shape (Ny, Nx)
    positions = np.vstack([T1.ravel(), T2.ravel()]) # Shape (2, Nx*Ny) for KDE

    # --- Calculate Reference PDF (Z_ref) from analytical function ---
    print(f"Calculating reference PDF on {Nx}x{Ny} grid...")
    log_Z_ref = np.zeros_like(T1)
    try:
        # Vectorized calculation if log_posterior_func supports it? Assumes not for safety.
        for r in range(Ny): # Loop over rows (y)
            for c in range(Nx): # Loop over columns (x)
                theta_ij = np.array([T1[r, c], T2[r, c]])
                log_Z_ref[r, c] = log_posterior_func(theta_ij)

        # Check for NaN/Inf in log posterior
        if not np.all(np.isfinite(log_Z_ref)):
             print("Warning: Non-finite values found in log_posterior_func output. Replacing with large negative number.")
             max_log = np.max(log_Z_ref[np.isfinite(log_Z_ref)]) if np.any(np.isfinite(log_Z_ref)) else 0
             log_Z_ref[~np.isfinite(log_Z_ref)] = max_log - 700 # Avoid exp underflow/overflow issues

        # Convert log posterior to posterior density & normalize
        log_Z_ref_shifted = log_Z_ref - np.max(log_Z_ref) # Shift for numerical stability
        Z_ref = np.exp(log_Z_ref_shifted)
        norm_factor_ref = np.sum(Z_ref) * dx * dy
        if norm_factor_ref > 1e-15:
            Z_ref /= norm_factor_ref
        else:
            print("Warning: Integral of reference PDF is close to zero. Check log_posterior_func or grid_range. Setting Z_ref to zeros.")
            Z_ref = np.zeros_like(Z_ref)

    except Exception as e:
        print(f"Error evaluating log_posterior_func on the grid: {e}. Skipping TV plot.")
        return

    # --- Calculate and Plot TV for each method ---
    print("Calculating TV distance trajectories...")
    fig3, ax3 = plt.subplots(figsize=(8, 6))

    for i, (history, title) in enumerate(zip(histories, titles)):
        tv_distances = []
        print(f"  Processing method: {title}...")
        iterations = range(len(history))

        for k in iterations:
            ensemble_k = history[k]
            tv_k = np.nan # Default to NaN

            # Check if ensemble is valid for KDE
            if ensemble_k.ndim == 2 and ensemble_k.shape[0] >= 2 and ensemble_k.shape[1] == 2:
                try:
                    kernel = gaussian_kde(ensemble_k.T)
                    Z_kde_k = np.reshape(kernel(positions).T, T1.shape)
                    norm_factor = np.sum(Z_kde_k) * dx * dy
                    if norm_factor > 1e-15: Z_kde_k /= norm_factor
                    else: Z_kde_k = np.zeros_like(Z_kde_k)
                    tv_k = np.sum(np.abs(Z_kde_k - Z_ref)) * dx * dy
                except (np.linalg.LinAlgError, ValueError) as e:
                     print(f"    Warning: KDE failed at iter {k} for {title}. Assigning NaN. Error: {e}")
            else:
                 if k == 0: print(f"    Warning: Invalid ensemble shape ({ensemble_k.shape}) for KDE at iter {k} for {title}. Assigning NaN.")

            tv_distances.append(tv_k)

        # Plot trajectory
        ax3.plot(iterations, tv_distances, label=title, marker='.', markersize=4, linestyle='-')

    # Customize the TV distance plot
    ax3.set_xlabel("Iteration")
    ax3.set_ylabel("TV Distance ($L_1$ Approx.)")
    ax3.set_title("Evolution of TV Distance from Reference Posterior")
    ax3.legend()
    ax3.grid(True, linestyle=':')
    ax3.set_ylim(bottom=0) # Ensure y-axis starts at 0
    fig3.tight_layout()
    plt.show()
```

```python

# === Likelihood Score Function ===
# Gradient of log p(y|x) w.r.t x = [x1, x2]
# log p(y|x) = C - 1/(2*sigma_R^2) * (y - (x1 - x2)^2)^2
# Nabla_x log p(y|x) = (2*(x1-x2)/sigma_R^2) * (y - (x1-x2)^2) * [1, -1]^T

def likelihood_score_2d(x, y_obs, sigma_R):
    """ Calculates Nabla_x log p(y_obs|x) for the 2D problem. """
    x1 = x[:, 0]
    x2 = x[:, 1]
    diff = x1 - x2
    g = diff**2
    common_factor = (2.0 * diff / sigma_R**2) * (y_obs - g)
    score_x1 = common_factor
    score_x2 = -common_factor
    # Stack into (batch_size, 2)
    score = torch.stack([score_x1, score_x2], dim=1)
    return score

# Make a partial function for the specific observation
likelihood_score_fn_2d = functools.partial(likelihood_score_2d, y_obs=y_obs_t, sigma_R=sigma_R_t)


# === Diffusion Model Setup (from Ref 1, adapted) ===
# Use the same diffusion parameters as Ref 1 for consistency? Or tune them? Let's start with Ref 1's.
sigma_diffusion = 25.0 # VP SDE parameter from Ref 1

# Define marginal_prob_std and diffusion_coeff using torch tensors
def marginal_prob_std_pytorch(t, sigma):
    """Compute the std dev of p(x(t)|x(0)) for VP SDE."""
    # Ensure t is tensor and on correct device
    if not isinstance(t, torch.Tensor):
        t = torch.tensor(t, device=device)
    t = t.to(device)
    # Use torch operations
    log_sigma = np.log(sigma)
    return torch.sqrt((torch.pow(sigma, 2 * t) - 1.) / (2. * log_sigma))

def diffusion_coeff_pytorch(t, sigma):
    """Compute the diffusion coefficient g(t) for VP SDE."""
    # Ensure t is tensor and on correct device
    if not isinstance(t, torch.Tensor):
        t = torch.tensor(t, device=device)
    t = t.to(device)
    return torch.pow(sigma, t)

marginal_prob_std_fn = functools.partial(marginal_prob_std_pytorch, sigma=sigma_diffusion)
diffusion_coeff_fn = functools.partial(diffusion_coeff_pytorch, sigma=sigma_diffusion)

# Re-define ScoreNet and related classes/functions from Ref 1
# (Assuming they are not already defined in the current scope if running sequentially)

class GaussianFourierProjection(nn.Module):
    """Gaussian random features for encoding time steps."""
    def __init__(self, embed_dim, scale=30.):
      super().__init__()
      self.W = nn.Parameter(torch.randn(embed_dim // 2) * scale, requires_grad=False)
    def forward(self, x):
      x_proj = x[:, None] * self.W[None, :] * 2 * np.pi
      return torch.cat([torch.sin(x_proj), torch.cos(x_proj)], dim=-1)

class ScoreNet(nn.Module):
    """A time-dependent score-based model adapted for lower dimensions."""
    def __init__(self, marginal_prob_std, x_dim=2 , hidden_depth=3, embed_dim=128, use_bn=False): # Reduced BN
        super().__init__()
        self.x_dim = x_dim
        self.hidden_depth = hidden_depth
        self.embed_dim = embed_dim
        self.use_bn = use_bn

        self.embed = nn.Sequential(GaussianFourierProjection(embed_dim=embed_dim),
                             nn.Linear(embed_dim, embed_dim))

        self.input = nn.Linear(x_dim, embed_dim)
        self.fc_all = nn.ModuleList([nn.Linear(embed_dim, embed_dim) for _ in range(self.hidden_depth)])
        self.output = nn.Linear(embed_dim, x_dim)

        if self.use_bn:
            # Reduced BN usage, apply only after act? Or skip? Let's skip BN for this low-dim problem initially.
            print("Warning: BatchNorm disabled for ScoreNet in this example.")
            self.use_bn = False
            # self.bn = nn.ModuleList([nn.BatchNorm1d(num_features=embed_dim) for _ in range(self.hidden_depth)])

        self.act = lambda x: F.leaky_relu(x, negative_slope=0.01) # Using LeakyReLU
        self.marginal_prob_std = marginal_prob_std

    def forward(self, x, t):
        embed = self.act(self.embed(t))
        h = self.act(self.input(x)) # Apply activation after input layer

        for i in range(self.hidden_depth):
            # Simple feedforward block with residual connection + embedding
            h_in = h
            h = self.fc_all[i](h)
            h = self.act(h + embed) # Add embedding before activation
            h = h + h_in # Residual connection (optional, common in deeper nets)
            # if self.use_bn: h = self.bn[i](h) # BN would go here if used

        h = self.output(h)
        # Normalize output by marginal std dev
        h = h / self.marginal_prob_std(t)[:, None]
        return h

def loss_fn(model, x, marginal_prob_std, eps=1e-5):
    """Loss function for training score models."""
    random_t = torch.rand(x.shape[0], device=x.device) * (1. - eps) + eps
    z = torch.randn_like(x)
    std = marginal_prob_std(random_t)
    perturbed_x = x + z * std[:, None]
    score = model(perturbed_x, random_t)
    loss = torch.mean(torch.sum((score * std[:, None] + z)**2, dim=1))
    return loss

# Dataset Class (from Ref 1)
class MyDataset(Dataset):
    def __init__(self, data_tensor):
      self.data_tensor = data_tensor
    def __len__(self):
      return self.data_tensor.shape[0]
    def __getitem__(self, idx):
      return self.data_tensor[idx]

# Training Function (from Ref 1)
def train_model(score_model, x_sample, lr=1e-3, batch_size=256, n_epochs=200, print_interval=20): # Adjust params
    """Trains the score model."""
    dataset = MyDataset(x_sample)
    # Reduce num_workers if issues arise
    data_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True, num_workers=0)

    optimizer = Adam(score_model.parameters(), lr=lr)
    train_loss_hist = [] # Use different name from outer scope if necessary
    score_model.train() # Set model to training mode

    for epoch in range(n_epochs):
      avg_loss = 0.
      num_items = 0
      for x_batch in data_loader:
        x_batch = x_batch.to(device)
        loss = loss_fn(score_model, x_batch, marginal_prob_std_fn)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        avg_loss += loss.item() * x_batch.shape[0]
        num_items += x_batch.shape[0]

      epoch_avg_loss = avg_loss / num_items
      train_loss_hist.append(epoch_avg_loss)
      if epoch % print_interval == 0 or epoch == n_epochs - 1:
          print(f'Epoch {epoch}: Loss = {epoch_avg_loss:.4f}')
    score_model.eval() # Set model back to evaluation mode
    return train_loss_hist


# Euler-Maruyama Sampler (from Ref 1, ensure tqdm is standard)
def Euler_Maruyama_sampler(score_prior,
                           marginal_prob_std,
                           diffusion_coeff,
                           score_likelihood=None, # Takes x, returns score
                           likelihood_weight_fn=None, # Takes t, returns weight h(t)
                           batch_size=512,
                           num_steps=500,
                           x_dim=2, # Specify x_dim
                           device='cpu',
                           eps=1e-3):
    """Generate samples using EM method."""
    t_init = torch.ones(batch_size, device=device)
    # Initial state is noise N(0, sigma_max^2 * I)
    init_x = torch.randn(batch_size, x_dim, device=device) * marginal_prob_std(t_init)[:, None]
    time_steps = torch.linspace(1., eps, num_steps, device=device)
    step_size = time_steps[0] - time_steps[1]
    x = init_x
    score_prior.eval() # Ensure model is in eval mode

    with torch.no_grad():
        for time_step in tqdm.tqdm(time_steps, desc="Sampling"):
            batch_time_step = torch.ones(batch_size, device=device) * time_step
            g = diffusion_coeff(batch_time_step)
            
            # Calculate combined score
            prior_score = score_prior(x, batch_time_step)
            if score_likelihood is not None:
                like_score = score_likelihood(x) # Assume it takes only x
                # Apply weighting based on time
                weight = 1.0 # Default weight if no function provided
                if likelihood_weight_fn is not None:
                    weight = likelihood_weight_fn(time_step) # Weight depends on current time_step
                
                # Simple score clipping (optional, might help stability)
                like_score = torch.clip(like_score, -100, 100)

                score = prior_score + like_score * weight
            else:
                score = prior_score

            # Optional: Clip total score or prior score if needed
            score = torch.clip(score, -100, 100) # Clip combined score

            # EM update step
            mean_x = x + (g**2)[:, None] * score * step_size
            x = mean_x + torch.sqrt(step_size) * g[:, None] * torch.randn_like(x)

    # Return the last mean_x (without final noise step)
    return mean_x

# Define a simple likelihood weight function h(tau) = 1 - tau (as suggested in slides)
def likelihood_weight_h(t_scalar, scale=1.0):
    # Ensure t_scalar is a scalar tensor or float
    if isinstance(t_scalar, torch.Tensor):
        t = t_scalar.item() # Get scalar value
    else:
        t = t_scalar
    # Simple linear decay from 1 at t=0 to 0 at t=1 (assuming T=1)
    # Use relu to ensure non-negative weight
    # Apply scaling factor if needed
    return scale * max(0.0, 1.0 - t)
```

## Train Prior Score Model

```python
# === Train Prior Score Model ===
print("\n--- Training Prior Score Model ---")
num_prior_samples = 20000 # Number of samples from N(mu0, P0)
prior_samples = mu0_t + torch.randn(num_prior_samples, 2, device=device) @ torch.linalg.cholesky(P0_t).T
# Or simpler if P0 is diagonal:
# prior_samples = mu0_t + torch.randn(num_prior_samples, 2, device=device) * torch.tensor(prior_std_dev, device=device)

# Instantiate the score model
score_prior_model = ScoreNet(marginal_prob_std=marginal_prob_std_fn, x_dim=2).to(device)

# Train the model
train_losses = train_model(score_prior_model, prior_samples,
                           lr=1e-3, batch_size=512, n_epochs=300, print_interval=50)

plt.figure()
plt.plot(train_losses)
plt.title("Prior Score Model Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.yscale('log')
plt.grid(True)
plt.show()
```

```python


# === Sample from Posterior using Trained Prior and Likelihood Score ===
print("\n--- Sampling from Posterior ---")
num_posterior_samples = 5000

# Define the likelihood weight function with any desired scaling
likelihood_weight_fn_scaled = functools.partial(likelihood_weight_h, scale=1.0) # Adjust scale if needed

posterior_samples = Euler_Maruyama_sampler(
    score_prior=score_prior_model,
    marginal_prob_std=marginal_prob_std_fn,
    diffusion_coeff=diffusion_coeff_fn,
    score_likelihood=likelihood_score_fn_2d, # Use the analytical likelihood score
    likelihood_weight_fn=likelihood_weight_fn_scaled, # Use the h(tau) function
    batch_size=num_posterior_samples,
    num_steps=1000, # Number of reverse steps
    x_dim=2,
    device=device,
    eps=1e-3
)

posterior_samples_np = posterior_samples.cpu().numpy()
print(f"Generated {posterior_samples_np.shape[0]} posterior samples.")
```
